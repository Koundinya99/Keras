```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Loading the dataset
df=pd.read_csv(r'C:\Users\medik\OneDrive\Desktop\CSV Files\pima-indians-diabetes.csv')
df
```

```python
#Splitting the data
X=df.iloc[:,0:8]
Y=df.iloc[:,8]
```

```python
#Importing keras libaries
from keras.models import Sequential
from keras.layers import Dense
```

```python
#Creating a model
model=Sequential()
#Adding input layer with 8 neurons and first hidden layer with 12 neurons and relu as activaion function
model.add(Dense(12,input_dim=8,activation='relu'))
#Adding second hidden layer with 8 neurons and relu as activaion function
model.add(Dense(8,activation='relu'))
#Adding Output layer with 1 neuron and sigmoid as activaion function
model.add(Dense(1,activation='sigmoid'))
#Compiling the model (Backpropagation)
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
#Training the model (via fit)
model.fit(X,pd.DataFrame(Y),epochs=150,batch_size=10,verbose=2)
```

```python
#Evaluating the model
score=model.evaluate(X,Y)
```

```python
#Predictions
predict=model.predict(X)
check=[round(X[0]) for X in predict]
df.rename(columns={'1':'Actual'})
df['Predicted']=pd.DataFrame(check)
df.rename(columns={'1':'Actual'},inplace=True)
pd.set_option('display.max_rows',None)
```
